```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()) )
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns = {'purchase_amount_(usd)': 'purchase_amount'})
```

```python
df.columns
```

***Create a column age_group to group the Customer by age***

```python
labels = ['Young Adults' , 'Adults' , 'Middle-aged' ,'Senior']
df['age_group'] = pd.qcut(df['age'],q=4,labels =labels)
```

```python
df[['age','age_group']].head(10)
```

***create purchase_frequency_days***


```python
# Standardize text in the column: lowercase & strip spaces
df['frequency_of_purchases'] = df['frequency_of_purchases'].str.strip().str.lower()

# Standardize mapping keys
frequency_mapping = {
    'fortnightly': 14,
    'weekly': 7,
    'monthly': 30,
    'quarterly': 90,
    'bi-weekly': 14,
    'annually': 365,
    'every 3 months': 90
}

# Map again
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```
